```python
from pylab import *

def f(x):
    return x**2 - 2

x = []
i = -3
while i < 4:
    x.append(i)
    i = i + 1
x = array(x)

plot(x,f(x),linestyle='--',marker='o')
show()
```

```python
from pylab import *

x = linspace(-3,3,1000)
y = x**2 - 2

plot(x,y)
show()
```

```python
from pylab import *

liste1 = [1,2,3]
liste2 = [5,4,3]

print("Listeaddisjon:", liste1 + liste2)

array1 = array(liste1)
array2 = array(liste2)

print("Arrayaddisjon:", array1+array2)
print("Arraymultiplikasjon:",array1*array2)
```

```python
from pylab import *

# Med lister
partall = []
for tall in range(0,101,2):
    partall.append(tall)

tall = 0
while tall <= 100:
    partall.append(tall)
    tall = tall + 2
print(partall)

# Med arrayer
partall = zeros(51)

for i in range(50):
    partall[i+1] = partall[i] + 2
print(partall)
```

```python
from pylab import *

x = linspace(-3,3,7)
f = x + 3
g = exp(0.01*x)
h = cos(x) - 1

plot(x,f,color='firebrick',marker='.',linestyle='--',label='f')
plot(x,g,color='hotpink',marker='*',label='g')
plot(x,h,color='navy',marker='>',label='h')
grid()
xlabel('tid (s)')
ylabel('Temperatur (celsius)')
legend()
show()
```
